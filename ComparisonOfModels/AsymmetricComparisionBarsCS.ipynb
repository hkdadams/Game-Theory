## Equilibrium Comparison with Bar Graphs for Cournot vs. Stackelberg 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# parameters for asymmetric costs:
alpha_d = 100
beta_d = 0.5
# marginal cost for Firm 1
c1 = 0.5   
# marginal cost for Firm 2
c2 = 0.1   

# -------------------------------
# Equilibrium Calculations:

# Cournot equilibrium (derived from best-response functions):
q1_C = (alpha_d + beta_d*(c2 - 2*c1)) / 3.0
q2_C = (alpha_d + beta_d*(c1 - 2*c2)) / 3.0
total_C = q1_C + q2_C
price_C = alpha_d - beta_d * total_C

# Stackelberg equilibrium (Firm 1 is leader, Firm 2 is follower):

q1_S = (alpha_d - beta_d*c1 + beta_d*c2) / 2.0
q2_S = (alpha_d - beta_d*c2 - q1_S) / 2.0
total_S = q1_S + q2_S
price_S = alpha_d - beta_d * total_S

# -------------------------------
# Profit Calculations:
# Profit = quantity * (market price - marginal cost)
profit_C1 = q1_C * (price_C - c1)
profit_C2 = q2_C * (price_C - c2)
profit_leader = q1_S * (price_S - c1)
profit_follower = q2_S * (price_S - c2)

```

```python
# AI Generated Bar Graphs

# -------------------------------
# Graph 1: Bar Chart for Individual Firm Outputs (Firm identity palette)

# pastel colours for firm identities:
# pastel lavender (Cournot for Firm 1)
firm1_color = "#d1c4e9"  
# pastel peach (Stackelberg for Firm 2)
firm2_color = "#ffccbc"  

labels = ['Firm 1', 'Firm 2']
x = np.arange(len(labels))
width = 0.35
# Firms' outputs under Cournot
cournot_outputs = [q1_C, q2_C]         
# Stackelberg: leader (Firm 1) and follower (Firm 2)
stackelberg_outputs = [q1_S, q2_S]       

plt.figure(figsize=(8,6))
plt.bar(x - width/2, cournot_outputs, width, label='Cournot', color=firm1_color)
plt.bar(x + width/2, stackelberg_outputs, width, label='Stackelberg', color=firm2_color)
plt.ylabel('Quantity Produced')
plt.title('Individual Firm Outputs: Cournot vs. Stackelberg')
plt.xticks(x, labels)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# -------------------------------
# Graph 2: Bar Chart for Total Market Quantities (Model palette)

# pastel colours for models:
# pastel purple
cournot_model_color = "#b39ddb"  
# pastel pink
stackelberg_model_color = "#f48fb1"

models = ['Cournot', 'Stackelberg']
total_quantities = [total_C, total_S]

plt.figure(figsize=(6,5))
plt.bar(models, total_quantities, color=[cournot_model_color, stackelberg_model_color])
plt.ylabel('Total Quantity Produced')
plt.title('Total Market Output Comparison')
plt.ylim(0, max(total_C, total_S)*1.1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# -------------------------------
# Graph 3: Bar Chart for Market Prices (Model palette)

prices = [price_C, price_S]

plt.figure(figsize=(6,5))
plt.bar(models, prices, color=[cournot_model_color, stackelberg_model_color])
plt.ylabel('Market Price')
plt.title('Market Price Comparison')
plt.ylim(0, alpha_d)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# -------------------------------
# Graph 4: Bar Chart for Individual Firm Profits (New firm model palette)
# pastel lavender for Cournot profits
cournot_profit_color = "#d1c4e9"   
# pastel peach for Stackelberg profits
stackelberg_profit_color = "#ffccbc" 

cournot_profits = [profit_C1, profit_C2]
stackelberg_profits = [profit_leader, profit_follower]

plt.figure(figsize=(8,6))
plt.bar(x - width/2, cournot_profits, width, label='Cournot', color=cournot_profit_color)
plt.bar(x + width/2, stackelberg_profits, width, label='Stackelberg', color=stackelberg_profit_color)
plt.ylabel('Profit')
plt.title('Individual Firm Profits: Cournot vs. Stackelberg')
plt.xticks(x, labels)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python

# results:

print("Cournot Equilibrium:")
print(f"  Firm 1: q1 = {q1_C:.2f}, Profit = {profit_C1:.2f}")
print(f"  Firm 2: q2 = {q2_C:.2f}, Profit = {profit_C2:.2f}")
print(f"  Total Quantity = {total_C:.2f}, Market Price = {price_C:.2f}\n")
```
